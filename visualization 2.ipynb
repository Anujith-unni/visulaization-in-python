```python
import seaborn as sns

data = sns.load_dataset('iris')

#style used as theme of graph
#for example if we want black
#graph with whitegrid then write "darkgrid"
sns.set_style("whitegrid")

#sepal_length, petal_lenght are iris
#feature data height used to define
#height of graph whereas hue store the 
#class of iris dataset
sns.FacetGrid(data, hue="species",
             height=6).map(plt.scatter,
                          'sepal_length',
                          'petal_length').add_legend()


```

```python
sns.heatmap(data.corr(), cmap = 'YlGnBu', linecolor = 'white', linewidths = 1)
```

```python
sns.heatmap(data.corr(), cmap = 'YlGnBu', linecolor = 'white', linewidths = 1,annot=True)
```

```python
data.corr(method='pearson')
```

```python
g = sns.pairplot(data,hue="species")
```

```python
# importing libraries
import seaborn as sb
# load data
df = sb.load_dataset('iris')
# use regplot
sb.regplot(x="sepal_length", #dependent variable
          y="petal_length", #independent variable
           ci=None,
          data=df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#load the tips dataset present by default in seaborn
tips=sns.load_dataset('tips')
sns.set_style('white')

#make a countplot
sns.countplot(x='sex', data=tips)
```

```python
tips=sns.load_dataset('tips')
sns.set_style('ticks')
sns.countplot(x='sex', data=tips, palette = 'deep')
```

```python
tips=sns.load_dataset('tips')
sns.set_style('darkgrid')

#make a countplot
sns.countplot(x='sex', data=tips)
```

```python
tips = sns.load_dataset('tips')
plt.figure(figsize=(3,5))
sns.countplot(x = 'sex', data=tips)
```
